```ocaml
let min e1 e2 = if e1 < e2 then e1 else e2;;
```

```ocaml
let max e1 e2 = if e1 > e2 then e1 else e2;;
```

```ocaml
let rec minmax l = match l with
    | [] -> (max_int, min_int)
    | e::q -> let mini, maxi = minmax q in
                ((min mini e), (max maxi e));;
```

```ocaml
minmax [1;5;3;7;0;3];;
```

```ocaml
let rec concat l1 l2 = match l1 with
    | [] -> l2
    | e::q -> e::(concat q l2);;
```

```ocaml
concat [2] [1];;
```

```ocaml
let rec partition p l = match l with
    | [] -> ([], [])
    | e::q -> let l1, l2 = (partition p q) in
                if e < p then ((e::l1), l2)
                else (l1, (e::l2));;
```

```ocaml
partition 1 [0;2;4;-1];;
```

```ocaml
let rec quicksort l = match l with
    | [] -> []
    | [e] -> [e]
    | e::q -> let l1, l2 = partition e l in
        concat (quicksort l1) (quicksort l2);;
```

```ocaml
quicksort [1;4;3;-7;-1;4;5;1;7];;
```
